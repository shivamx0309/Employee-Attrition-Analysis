# Employee Attrition Analysis

## Objective
To analyze employee data and identify key factors responsible for employee attrition in an organization.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cleaned_attrition.csv")
df.head()
```

# 1:-
## This represents the percentage of employees leaving the organization

```python
attrition_rate = df["Attrition"].mean() * 100
print(f"Overall Attrition Rate: {attrition_rate:.2f}%")
```

## The imbalance between retained and leaving employees indicates retention challenges

```python
sns.countplot(x="Attrition", data=df)
plt.title("Employee Attrition Distribution")
plt.show()
```

# 2:-
## Some departments show higher attrition, suggesting role-specific stress or growth issues.

```python
sns.countplot(x="Department", hue="Attrition", data=df)
plt.xticks(rotation=30)
plt.title("Department-wise Attrition")
plt.show()
```

# 3:-
## Some departments show higher attrition, suggesting role-specific stress or growth issues.

```python
sns.countplot(x="Department", hue="Attrition", data=df)
plt.xticks(rotation=30)
plt.title("Department-wise Attrition")
plt.show()
```

# 4:-
## Lower-paid employees are more likely to leave, indicating compensation impact.

```python
sns.boxplot(x="Attrition", y="MonthlyIncome", data=df)
plt.title("Monthly Income vs Attrition")
plt.show()
```

# 5:-
## Employees working overtime tend to leave more frequently due to workload imbalance

```python
sns.countplot(x="OverTime", hue="Attrition", data=df)
plt.title("Overtime Impact on Attrition")
plt.show()
```

# 6:-
## Lower job satisfaction strongly correlates with higher attrition.

```python
sns.countplot(x="JobSatisfaction", hue="Attrition", data=df)
plt.title("Job Satisfaction vs Attrition")
plt.show()
```

# 7:-
## Employees with shorter tenure are more likely to leave early

```python
sns.boxplot(x="Attrition", y="YearsAtCompany", data=df)
plt.title("Tenure vs Attrition")
plt.show()
```

# 8:-
## â€œCorrelation highlights key numerical drivers influencing attrition.â€

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=["int64", "float64"])

plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

## Why did correlation heatmap fail?

## Correlation works only on numerical features. I filtered numeric columns using select_dtypes() before plotting the heatmap.
