Step:- 5 Outlier check (salary, age, years)

```python
plt.hist(df["MonthlyIncome"], bins=30)
plt.title("Monthly Income Distribution")
plt.show()
```

“I usually detect outliers using,histogram,”

```python
Q1 = df["MonthlyIncome"].quantile(0.25)
Q3 = df["MonthlyIncome"].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print("Lower bound:", lower)
print("Upper bound:", upper)

outliers = df[(df["MonthlyIncome"] < lower) | (df["MonthlyIncome"] > upper)]
print("Outliers count:", outliers.shape[0])
```

I detected outliers in Monthly Income using the IQR method.
However, I retained them because high salary values represent valid senior employees rather than data errors.
